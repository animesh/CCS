```python
#Generated libraries dont contain modifications
measured = measured[measured["Modifications"]=="Unmodified"]
#Modified sequence is therefore just the _sequence_
measured["Modified sequence"] = "_" + measured["Sequence"] + "_"
```

```python
measured_subset = measured[double.columns]
```

```python
#merge all the dataframes
merged = double.copy()
merged.rename({'Calibrated 1/K0':'Calibrated 1/K0_double'}, axis=1, inplace=True)
merged.loc[:,'Calibrated 1/K0_gauss'] = gauss['Calibrated 1/K0'].values
merged.loc[:, 'Calibrated 1/K0-daniel_ch2'] = daniel_ch2['Calibrated 1/K0'].values
```

```python
merged = pd.merge(merged, pred[['Modified sequence', 'Charge', 'Calibrated 1/K0']], how='left', on=['Modified sequence', 'Charge'], suffixes=[None, 'pred'])
merged.rename({'Calibrated 1/K0':'Calibrated 1/K0_pred'}, axis=1, inplace=True)
merged = pd.merge(merged, measured[['Modified sequence', 'Charge', 'Calibrated 1/K0']], how='left', on=['Modified sequence', 'Charge'], suffixes=[None, 'measured'])
merged.rename({'Calibrated 1/K0':'Calibrated 1/K0_measured'}, axis=1, inplace=True)
```

```python
merged_valid = merged[~merged['Calibrated 1/K0_measured'].isna()]
```

```python
merged_valid.loc[:,'error_double'] = np.abs(merged_valid['Calibrated 1/K0_double'] - merged_valid['Calibrated 1/K0_measured'])
merged_valid.loc[:,'error_gauss'] = np.abs(merged_valid['Calibrated 1/K0_gauss'] - merged_valid['Calibrated 1/K0_measured'])
merged_valid.loc[:,'error_daniel_ch2'] = np.abs(merged_valid['Calibrated 1/K0_pred'] - merged_valid['Calibrated 1/K0_measured'])
```

```python
best_double = merged_valid.sort_values(['Modified sequence', 'Charge', 'error_double'], ascending=True).groupby(['Modified sequence', 'Charge'])['Calibrated 1/K0_double'].first()
best_gauss = merged_valid.sort_values(['Modified sequence', 'Charge', 'error_gauss'], ascending=True).groupby(['Modified sequence', 'Charge'])['Calibrated 1/K0_gauss'].first()
best_daniel_ch2 = merged_valid.sort_values(['Modified sequence', 'Charge', 'error_daniel_ch2'], ascending=True).groupby(['Modified sequence', 'Charge'])['Calibrated 1/K0-daniel_ch2'].first()
```

```python
best_double.shape, best_gauss.shape, best_daniel_ch2.shape
```

```python
#merge pred, measured, best_double, best_gauss
merged = pd.merge(best_double, best_gauss, how='inner', on=['Modified sequence', 'Charge'], suffixes=['_double', '_gauss'])
merged = pd.merge(merged, best_daniel_ch2, how='inner', on=['Modified sequence', 'Charge'])
merged = pd.merge(merged, pred[['Modified sequence', 'Charge', 'Calibrated 1/K0']], how='inner', on=['Modified sequence', 'Charge'], suffixes=[None, '_pred'])
merged = pd.merge(merged, measured[['Modified sequence', 'Charge', 'Calibrated 1/K0']], how='inner', on=['Modified sequence', 'Charge'], suffixes=[None, '_measured'])
```

```python
merged.rename({'Calibrated 1/K0':'Calibrated 1/K0_pred'}, axis=1, inplace=True)
```

```python
merged = merged.loc[:, ~merged.columns.duplicated()].copy()
```

```python
merged
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8,6))
for column, name in zip(['Calibrated 1/K0_double', 'Calibrated 1/K0_gauss', 'Calibrated 1/K0_pred', 'Calibrated 1/K0-daniel_ch2'], ['dynamic binning', 'gaussian - most intense', 'state of the art', 'daniel ch2']):
    rel_err = (merged[column] - merged['Calibrated 1/K0_measured'])/merged['Calibrated 1/K0_measured'] * 100
    mad = median_abs_deviation(rel_err, scale=1/1.4826)
    ax.hist(rel_err, bins=range(-100,100), alpha=0.5, label=f"{name} - MAD: {mad:.3f}")
    print(name, rel_err[rel_err.abs() > 10].shape)
ax.set_title("Relative error distribution")
ax.set_xlabel('Relative error (%)')
ax.set_ylabel('Frequency')
plt.tight_layout()
plt.legend()
```

```python
#create distribution plot per charge
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(20,5))

for charge in [2,3]:
    merged_charge = merged[merged['Charge'] == charge]
    for column in ['Calibrated 1/K0_double', 'Calibrated 1/K0_gauss', 'Calibrated 1/K0_pred']:
        rel_err = (merged_charge[column] - merged_charge['Calibrated 1/K0_measured'])/ merged_charge['Calibrated 1/K0_measured'] * 100
        mad = median_abs_deviation(rel_err, scale=1/1.4826)
        axs[charge-2].hist(rel_err, bins=range(-20,20), alpha=0.5, label=f"{column} - MAD: {mad:.3f}")
    axs[charge-2].set_title(f"Relative error distribution - Charge {charge}")
    axs[charge-2].set_xlabel('Relative error (%)')
    axs[charge-2].set_ylabel('Frequency')
    axs[charge-2].legend()
```

```python
#plot histogram of charges for all the dataframes
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15,5))
labels = ['double', 'gauss', 'pred']
for i, df in enumerate([double, gauss, pred]):
    axs[i].hist(df['Charge'], bins=20, log=True)
    axs[i].set_title(f"Charge distribution - {labels[i]}")
    axs[i].set_xlabel('Charge')
    axs[i].set_ylabel('Frequency')
```

```python
#navigate throw the folders and plot the results

results = {}
tables = ['evidence', 'peptides', 'proteinGroups','msms']
names = ["birnn empirical","xgb single value","birnn single value","birnn geom. fit","no 1/k0","xgb geom. fit",  "xgb empirical"]
for table in tables:
    print(table)
    runs = {}
    j=0
    for i, file in enumerate(glob.glob(f'{root_path}/test/fdr1/60SPD_py8_combined*/txt/{table}.txt', recursive=True)):
        #run = file.split('/')[-3][9:]
        run = names[j]
        j+=1
        print(file, run)
        try:
            df = pd.read_csv(file, sep='\t')
            print(df.shape)
        except pd.errors.EmptyDataError:
            print(f"Empty file: {file}")
            continue
        runs[run] = df
    results[table] = runs
```

```python
plt.style.use('science')
```

```python
import matplotlib.pyplot as plt
```

```python
# %%
#do a bar plot per table with all the runs with the length of the table
fig, axs = plt.subplots(nrows=2,ncols=2, figsize=(20, 15))
ax = axs.flatten()
n = len(runs)
colors = list(plt.cm.jet(np.linspace(0,1,n)))
for i, table in enumerate(tables):
    #run_index = list(results[table].keys())  # Use the keys directly
    run_index = ["rnn empirical","xgb empirical","rnn geom. fit","xgb geom. fit", "rnn single value","xgb single value", "no 1/k0"]
    tables_length = [len(results[table][idx]) for idx in run_index]
    #run_index = ["parameter-free model","xgboost with counts", "physical model", "Meier et al.", "gaussian mixture", "no 1/K0 prediction"]
    data = pd.DataFrame({'Run': run_index, 'Length': tables_length})
    bar = sns.barplot(x='Run', y='Length', data=data, ax=ax[i], palette=sns.color_palette("rainbow", len(runs)))
    ax[i].set_title(table, fontsize=16)
    ax[i].set_xlabel('Run', fontsize=16)
    ax[i].set_ylabel('Identifications', fontsize=16)
    ax[i].set_xticklabels(ax[i].get_xticklabels(), rotation=20, horizontalalignment='right', fontsize=14)    
    # Add the height of the bars as text on top of each bar
    for idx, length in enumerate(tables_length):
        ax[i].text(idx, length + 0.05, length, ha='center', va='bottom')
    plt.suptitle('Length of the tables per run without allignment - parameter-free separatoin\n', fontsize=20)
plt.tight_layout()
```
